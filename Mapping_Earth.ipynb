# Embedding Images onto Surfaces

```python
#Import needed tools
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Embedding the map of the earth onto cone-like surfaces
#### Image taken from Google and can be found here: http://map.vbgood.com/world_relief_map.htm

```python
#The image was saved locally.
img=mpimg.imread('earth.png')
```

```python
#Show surfaces in Notebook
%matplotlib inline
```

```python
fig=plt.figure()

plt.axis('off')
imgplot = plt.imshow(img)
#fig.savefig('Earth4.png')
```

```python
#Creating cone

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

r = np.linspace(0, 1.5, 220)
p = np.linspace(0, 2*np.pi, 220)
R, P = np.meshgrid(r, p)
Z = R

X, Y = R*np.cos(P), R*np.sin(P)

ax.grid(False)
plt.axis('off')
ax.plot_surface(X, Y, Z, facecolors=img,cstride=1,rstride=1)

ax.set_zlim(0, 1)

ax.view_init(elev=1.0, azim=115.0)
plt.show()
#fig.savefig('Earth3.png')
```

```python
def fun(x, y):
    return 2 * np.sqrt((x**2)/9 + (y**2)/9)
fig=plt.figure()
ax = Axes3D(fig)


x = y = np.arange(-2.0, 2.0, 0.02)

X, Y = np.meshgrid(x, y)

zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
ax.grid(False)
plt.axis('off')
ax.plot_surface(X, Y, Z,facecolors=img, cstride=1,rstride=1)
ax.view_init(elev=5.0, azim=309.0)
#plt.savefig('Earth1.png')
```
